```python
import tensorflow as tf
from tensorflow.keras import regularizers
import matplotlib.pyplot as plt

# Asegúrate de que train_X es un NumPy array
train_X = train_X.to_numpy() if isinstance(train_X, pd.DataFrame) else train_X

# Crea el modelo secuencial
modelf = tf.keras.Sequential([
    tf.keras.layers.Normalization(),  # Capa de normalización de los datos
    tf.keras.layers.Dense(units=256, activation='relu', kernel_regularizer=regularizers.l2(0.005)),  # Capa densa con más unidades
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),  # Dropout ajustado a 0.3
    tf.keras.layers.Dense(units=128, activation='relu', kernel_regularizer=regularizers.l2(0.005)),  # Otra capa densa
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),  # Dropout ajustado
    tf.keras.layers.Dense(units=64, activation='relu', kernel_regularizer=regularizers.l2(0.005)),
    tf.keras.layers.Dense(units=1)  # Capa de salida para regresión
])

# Ajustar la capa de normalización a los datos
modelf.layers[0].adapt(train_X)  # Ajusta la capa de normalización a los datos de entrada

# Compilar el modelo con un optimizador ajustado y MSE como la pérdida
modelf.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),  # Optimizador Adam con tasa de aprendizaje ajustada
    loss='mean_squared_error',  # Pérdida MSE para regresión
    metrics=['mae', 'mse']  # Métricas de evaluación
)

# Entrenar el modelo
history = modelf.fit(
    train_X,
    train_y,
    epochs=30,  # Número de épocas aumentado
    validation_split=0.2,  # Usar el 20% de los datos para validación
    batch_size=256,  # Tamaño de lote ajustado
    callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)],  # Early stopping
    verbose=1  # Imprimir información durante el entrenamiento
)

# Evaluar el modelo en el conjunto de prueba
test_loss, test_mae, test_mse = modelf.evaluate(test_X, test_y)
print(f"Test Loss: {test_loss}, Test MAE: {test_mae}, Test MSE: {test_mse}")

# Graficar los resultados
fig, axs = plt.subplots(3, 1, figsize=(10, 18))  # Tres subgráficas

# Pérdida (Loss)
axs[0].plot(history.history['loss'], label='Pérdida de Entrenamiento', color='blue')
axs[0].plot(history.history['val_loss'], label='Pérdida de Validación', color='orange', linestyle='--')
axs[0].set_xlabel('Época')
axs[0].set_ylabel('Pérdida')
axs[0].legend()
axs[0].set_title('Pérdida durante el entrenamiento')

# MAE (Error Absoluto Medio)
axs[1].plot(history.history['mae'], label='MAE de Entrenamiento', color='green')
axs[1].plot(history.history['val_mae'], label='MAE de Validación', color='red', linestyle='--')
axs[1].set_xlabel('Época')
axs[1].set_ylabel('MAE')
axs[1].legend()
axs[1].set_title('Error Absoluto Medio (MAE) durante el entrenamiento')

# MSE (Error Cuadrático Medio)
axs[2].plot(history.history['mse'], label='MSE de Entrenamiento', color='purple')
axs[2].plot(history.history['val_mse'], label='MSE de Validación', color='brown', linestyle='--')
axs[2].set_xlabel('Época')
axs[2].set_ylabel('MSE')
axs[2].legend()
axs[2].set_title('Error Cuadrático Medio (MSE) durante el entrenamiento')

# Mostrar las gráficas
plt.tight_layout()
plt.show()
```

SEGUNDO MODELO DE PRUEBA MEJORA LAS MÉTRICAS

```python
import tensorflow as tf
from tensorflow.keras import regularizers
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

# Asegúrate de que train_X es un NumPy array
train_X = train_X.to_numpy() if isinstance(train_X, pd.DataFrame) else train_X
test_X = test_X.to_numpy() if isinstance(test_X, pd.DataFrame) else test_X

# Escalar los datos utilizando StandardScaler
scaler = StandardScaler()
train_X = scaler.fit_transform(train_X)
test_X = scaler.transform(test_X)

# Crear el modelo secuencial
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(train_X.shape[1],)),  # Capa de entrada explícita
    tf.keras.layers.Dense(units=512, activation='relu', kernel_regularizer=regularizers.l2(0.01)),  # Más unidades y regularización L2 ajustada
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.4),  # Dropout ajustado a 0.4 para mayor regularización
    tf.keras.layers.Dense(units=256, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(units=128, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    tf.keras.layers.Dense(units=1)  # Capa de salida para regresión
])

# Compilar el modelo con un optimizador ajustado y MSE como la pérdida
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0003),  # Optimización más conservadora
    loss='mean_squared_error',  # Pérdida MSE para regresión
    metrics=['mae', 'mse']  # Métricas de evaluación
)



# Entrenar el modelo
history = model.fit(
    train_X,
    train_y,
    epochs=100,  # Aumento de épocas para permitir mayor aprendizaje
    validation_split=0.2,  # Usar el 20% de los datos para validación
    batch_size=256,  # Tamaño de lote ajustado
    callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=20)],  # Early stopping más conservador
    verbose=1  # Imprimir información durante el entrenamiento
)

# Evaluar el modelo en el conjunto de prueba
test_loss, test_mae, test_mse = model.evaluate(test_X, test_y)
print(f"Test Loss: {test_loss}, Test MAE: {test_mae}, Test MSE: {test_mse}")

# Calcular R²
predictions = model.predict(test_X)
r2 = r2_score(test_y, predictions)
print(f"R-squared: {r2}")

# Graficar los resultados
fig, axs = plt.subplots(3, 1, figsize=(10, 18))  # Tres subgráficas

# Pérdida (Loss)
axs[0].plot(history.history['loss'], label='Pérdida de Entrenamiento', color='blue')
axs[0].plot(history.history['val_loss'], label='Pérdida de Validación', color='orange', linestyle='--')
axs[0].set_xlabel('Época')
axs[0].set_ylabel('Pérdida')
axs[0].legend()
axs[0].set_title('Pérdida durante el entrenamiento')

# MAE (Error Absoluto Medio)
axs[1].plot(history.history['mae'], label='MAE de Entrenamiento', color='green')
axs[1].plot(history.history['val_mae'], label='MAE de Validación', color='red', linestyle='--')
axs[1].set_xlabel('Época')
axs[1].set_ylabel('MAE')
axs[1].legend()
axs[1].set_title('Error Absoluto Medio (MAE) durante el entrenamiento')

# MSE (Error Cuadrático Medio)
axs[2].plot(history.history['mse'], label='MSE de Entrenamiento', color='purple')
axs[2].plot(history.history['val_mse'], label='MSE de Validación', color='brown', linestyle='--')
axs[2].set_xlabel('Época')
axs[2].set_ylabel('MSE')
axs[2].legend()
axs[2].set_title('Error Cuadrático Medio (MSE) durante el entrenamiento')

# Mostrar las gráficas
plt.tight_layout()
plt.show()
```
